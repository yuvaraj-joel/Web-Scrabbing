#### Import The Libraries

```python
#pandas
import pandas as pd
#numpy
import numpy as np
#seaborn
import seaborn as sns 
#matplotlib
import matplotlib.pyplot as plt
#beatifulSoup
from bs4 import BeautifulSoup
#requests
import requests
#regEx
import re
```

```python
url="https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
req=requests.get(url)
```

```python
req
```

```python
soup=BeautifulSoup(req.content)
```

```python
arr=[]
for i in soup.find_all('td'):
    arr.append(str(i))
arr[0]
```

```python
re.sub('<td>.*">|<.*>',"",arr[0])
```

```python
re.sub('<td>.*">|<td>|\n|<\td>|<.*>',"",arr[0])
```

```python
Film=[]
Year=[]
Award=[]
Nomination=[]
count=0
for i in soup.find_all("td"):
    i=re.sub('<td>.*">|<td>|\n|<\td>|<.*>',"",str(i))
    count+=1
    if(count==1):
        Film.append(i)
    elif(count==2):
        Year.append(i)
    elif(count==3):
        Award.append(i)
    elif(count==4):
        Nomination.append(i)
        count=0
```

```python
Film
```

```python
oscar=pd.DataFrame({"Film":Film[:1360],"Awards":Award[:1360],"Year":Year[:1360],"Nominations":Nomination[:1360]})
```
